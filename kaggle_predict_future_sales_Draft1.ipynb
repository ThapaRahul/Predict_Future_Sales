```
import numpy as np
def downcast(df):
  df1 = df.copy()
  for item in df1.columns:
    if (df1[item].dtype == 'float64'):
      df1[item] = df1[item].astype(np.float32)
    elif (df1[item].dtype == "int64"):
      df1[item] = df1[item].astype(np.int16)
  return df1
```
